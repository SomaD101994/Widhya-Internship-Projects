### 4. Data Cleaning

```python
#Handling Missing Values
df.isnull().sum()
```

```python
#Replacing NaN Values with mean Value
df["btc_total_bitcoins"].fillna(df["btc_total_bitcoins"].mean(),inplace=True)
df["btc_trade_volume"].fillna(df["btc_trade_volume"].mean(),inplace=True)
df["btc_blocks_size"].fillna(df["btc_blocks_size"].mean(),inplace=True)
df["btc_median_confirmation_time"].fillna(df["btc_median_confirmation_time"].mean(),inplace=True)
df["btc_difficulty"].fillna(df["btc_difficulty"].mean(),inplace=True)
df["btc_transaction_fees"].fillna(df["btc_transaction_fees"].mean(),inplace=True)
df.isnull().sum()
```

```python
df.info()
```

### 5. Preparing the dataset

+ **Here we will divide the dataset into attributes (input) and labels (output). Then we will split the dataset into two parts - Testing data and Training data**

```python
X = np.array(df[['btc_market_cap','btc_n_transactions','btc_miners_revenue','btc_cost_per_transaction','btc_difficulty','btc_hash_rate','btc_cost_per_transaction_percent']])
Y = np.array(df[['btc_market_price']])
print(X.shape)
print(Y.shape)
```

### 6. Splitting dataset into training and test set

```python
#spliting the data into training and testing data. The ratio of this data is 20% (Test data) and 80% (Train Data)
X_train,X_test,y_train,y_test = train_test_split(X,Y,random_state = 0,test_size=0.2)
```

### 7. Traning the Algorithm

```python
print("X train.shape =", X_train.shape)
print("y train.shape =", y_train.shape)
print("X test.shape  =", X_test.shape)
print("y test.shape  =", y_test.shape)
```

```python
linreg=linear_model.LinearRegression()
```

```python
# Here we use fit function to tell the algorithm on which data to work
linreg.fit(X_train,y_train)
```

```python
#plotting the Regression Line (Y = MX + C)
Y0 = linreg.intercept_ + linreg.coef_*X_train
```

### 8. Predictions Making

```python
#predicting the Scores for test data
y_predicted=linreg.predict(X_test)
print(y_predicted)
```

### 9. Evaluating the Model

```python
# Finding the accuracy of the model
import math
from sklearn.metrics import mean_squared_error
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_predicted))
print("Mean Squared Error:", metrics.mean_squared_error(y_test,y_predicted))
print("Root Mean Squared Error:", math.sqrt(mean_squared_error(y_test,y_predicted)))
```
